```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
url = '/content/insurance.csv'
data = pd.read_csv(url)
data.sample(5)
```

```python
data.info()
```

```python
print(data['sex'].value_counts())
sns.countplot(x='sex', data=data)
```

```python
print(data['smoker'].value_counts())
sns.countplot(x='smoker', data=data)
```

```python
print(data['region'].value_counts())
sns.countplot(x='region', data=data)
```

```python
corr = data.corr()
sns.heatmap(corr,cmap='Wistia', annot=True)
```

```python
sns.pairplot(data)
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['charges'], data['charges']**2]) . T
model.fit(X, y = data['age'])
print('Intercept cua mo hinh :',model.intercept_)
print('Coeffient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh :',model.score(X,y = data['age']))
```

```python
import numpy as np
range = np.arange(0, 60000, 10)
X_range = pd.DataFrame([range, range**2]).T
y_predict = model.predict(X=X_range)
data.plot(kind='scatter', x='charges', y='age', figsize=(9,9), color='black')
plt.plot(range, y_predict, color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['charges'], data['charges']**2, data['charges']**3]) . T
model.fit(X, y = data['age'])
print('Intercept cua mo hinh :',model.intercept_)
print('Coeffient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh :',model.score(X,y = data['age']))
import numpy as np
range = np.arange(0, 60000, 10)
X_range = pd.DataFrame([range, range**2, range**3]).T
y_predict = model.predict(X=X_range)
data.plot(kind='scatter', x='charges', y='age', figsize=(9,9), color='black')
plt.plot(range, y_predict, color='blue')
```
